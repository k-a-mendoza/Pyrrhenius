```python
from pyrrhenious import database
import os
csv = 'publication_database.csv'

ecdata = database.Database(csv)
for pub in ecdata.database['Publication ID'].unique():
    try:
        file = f'publication_images/{pub}'
        os.mkdir(file)
        print(f'made {file}')
    except:
        pass
```

```python
eqs = [x.split('+') for x in ecdata.database['Eq ID'].unique()]
mechanisms = sorted(list(set([item.strip() for sublist in eqs for item in sublist])))
for x in mechanisms:
    print(x)
```

```python
for mechanism in mechanisms:
    print('='*20)
    print(mechanism)
    print('*'*20)
    values= ecdata.database[ecdata.database['Eq ID'].str.contains(mechanism)]['Equation Form']
    for x in values:
        print(x)
```
