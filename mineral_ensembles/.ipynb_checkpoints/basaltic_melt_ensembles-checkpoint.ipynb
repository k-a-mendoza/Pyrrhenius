# Exploring Pyrrhenious's Melt Models Database Correspondence

Here we look at a basaltic melt's parameterization correspondence of:

 * Pommier A. et al., 2010. Time-dependent changes of the electrical conductivity of basaltic melts with redox state
 
 * Ni H. et al., 2011. Electrical conductivity of hydrous basaltic melts: implications for partial melting in the upper mantle
 
 * Sifr√© D. et al., 2014. The electrical conductivity during incipient melting in the oceanic low-velocity zone

 * Dai et al., 2015. Electrical conductivity of hydrous natural basalts at high temperatures and pressures

```python
import matplotlib.cm as cm
P = 0 # GPa
t = np.linspace(773,2273,num=100)
logfo2 = pyrutils.calc_QFM(t,P)
phases = ecdatabase.get_phases()
model_names = ecdatabase.get_model_list_for_phase('basaltic melt')
fix, ax = plt.subplots(figsize=(7,7))
for i, model in enumerate(filter(lambda x : 'pom' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range = np.linspace(ec_model_row['temp_min'],
                                    ec_model_row['temp_max'],num=100,dtype=float).ravel()
    logfo2 = pyrutils.calc_QFM(temperature_range,P)
    pressure = ec_model_row['pressure_average'].astype(float).values
        
    if pressure< 0.001:
        pressure = 0

    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=P,logfo2=logfo2)),
            linewidth=2,label=model+f' QFM: {pressure} GPa',color='black')
    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=P,logfo2=logfo2-3)),
            linewidth=2,label=model+f' QFM-3: {pressure} GPa',color='black')
    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=P,logfo2=logfo2+3)),
            linewidth=2,label=model+f' QFM+3: {pressure} GPa',color='black')
for i, model in enumerate(filter(lambda x : 'ni' in x, model_names)):
    
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range = np.linspace(ec_model_row['temp_min'],
                                    ec_model_row['temp_max'],num=100,dtype=float).ravel()
    pressure = ec_model_row['pressure_average'].astype(float).values[0]
    print(pressure)

    if pressure< 0.001:
        pressure = 0
    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=P,Cw=0)),
            linewidth=2,label=model+f' 0% water: {pressure} GPa',color='red')
    
for i, model in enumerate(filter(lambda x : 'si' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range = np.linspace(ec_model_row['temp_min'],
                                    ec_model_row['temp_max'],num=100,dtype=float).ravel()
    pressure = ec_model_row['pressure_average'].astype(float).values[0]
    if pressure< 0.001:
        pressure = 0
    logfo2 = pyrutils.calc_QFM(temperature_range,P)
    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=P,Cw=0,co2=0)),
            linewidth=2,label=model+f' 0% water& CO2: {pressure} GPa',color='green')
    
for i, model in enumerate(filter(lambda x : 'dai' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range = np.linspace(ec_model_row['temp_min'],
                                    ec_model_row['temp_max'],num=100,dtype=float).ravel()
    pressure = ec_model_row['pressure_average'].astype(float).values[0]
    if pressure< 0.001:
        pressure = 0
    ax.plot(1e4/temperature_range,
            np.log10(ec_model.get_conductivity(T=temperature_range,P=2,Cw=0.62e4,co2=0)),
            linewidth=2,label=model+f' 0.62% water& CO2: 2 GPa',color='blue')
plt.legend(loc='lower left')
plt.title('Dry Basalt Melt Conductivity Trends')
plt.xlabel('1e4/T')
plt.ylabel('log10(sigma)')
```

```python
temperature_range_extended = np.linspace(1200,2000,num=802,dtype=float).ravel()
pom_values1 = np.zeros(len(temperature_range_extended))
pom_values1 = np.zeros(len(temperature_range_extended))
pom_values1 = np.zeros(len(temperature_range_extended))

for i, model in enumerate(filter(lambda x : 'pom' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range_1 = np.arange(int(ec_model_row['temp_min'].values[0]),
                                    int(ec_model_row['temp_max'].values[0])+1,1,dtype=float).ravel()
    logfo2 = pyrutils.calc_QFM(temperature_range_1,0)
    ec1 = ec_model.get_conductivity(T=temperature_range_1,P=0,logfo2=logfo2)
    ec2 = ec_model.get_conductivity(T=temperature_range_1,P=0,logfo2=logfo2-3)
    ec3 = ec_model.get_conductivity(T=temperature_range_1,P=0,logfo2=logfo2+3)
    logfo2 = pyrutils.calc_QFM(temperature_range_extended,0)
    ec1_ext = ec_model.get_conductivity(T=temperature_range_extended,P=0,logfo2=logfo2)
    ec2_ext = ec_model.get_conductivity(T=temperature_range_extended,P=0,logfo2=logfo2-3)
    ec3_ext = ec_model.get_conductivity(T=temperature_range_extended,P=0,logfo2=logfo2+3)
    
    df1 = pd.Series(ec1,index = temperature_range_1,name='pom qfm')
    df2 = pd.Series(ec2,index = temperature_range_1,name='pom qfm-3')
    df3 = pd.Series(ec3,index = temperature_range_1,name='pom qfm+3')
    pom_df_within = pd.concat([df1,df2,df3],axis=1)
    df1 = pd.Series(ec1_ext,index = temperature_range_extended,name='pom qfm')
    df2 = pd.Series(ec2_ext,index = temperature_range_extended,name='pom qfm-3')
    df3 = pd.Series(ec3_ext,index = temperature_range_extended,name='pom qfm+3')
    pom_df_outside = pd.concat([df1,df2,df3],axis=1)
    
for i, model in enumerate(filter(lambda x : 'ni' in x, model_names)):
    
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range_2 = np.arange(int(ec_model_row['temp_min'].values[0]),
                                    int(ec_model_row['temp_max'].values[0])+1,dtype=float).ravel()
    ec4 = ec_model.get_conductivity(T=temperature_range_2,P=0,Cw=0)
    ec4_ext = ec_model.get_conductivity(T=temperature_range_extended,P=0,Cw=0)
    ni_within = pd.Series(ec4,index = temperature_range_2,name='Ni dry')
    ni_outside = pd.Series(ec4_ext,index = temperature_range_extended,name='Ni dry')
    
for i, model in enumerate(filter(lambda x : 'si' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range_3 = np.arange(int(ec_model_row['temp_min'].values[0]),
                                    int(ec_model_row['temp_max'].values[0])+1,dtype=float).ravel()
    ec5    = ec_model.get_conductivity(T=temperature_range_3,P=0,Cw=0,co2=0)
    ec5_ext = ec_model.get_conductivity(T=temperature_range_extended,P=0,Cw=0,co2=0)
    sif_within  = pd.Series(ec5,index = temperature_range_3,name='Sif dry')
    sif_outside = pd.Series(ec5_ext,index = temperature_range_extended,name='Sif dry')
    
for i, model in enumerate(filter(lambda x : 'dai' in x, model_names)):
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model     = ecdatabase.get_model(model)
    temperature_range_3 = np.arange(int(ec_model_row['temp_min'].values[0]),
                                    int(ec_model_row['temp_max'].values[0])+1,dtype=float).ravel()
    ec5    = ec_model.get_conductivity(T=temperature_range_3,P=0.5,Cw=0.62,co2=0)
    ec5_2    = ec_model.get_conductivity(T=temperature_range_3,P=2,Cw=0.62,co2=0)
    ec5_ext_1 = ec_model.get_conductivity(T=temperature_range_extended,P=0,Cw=0.62,co2=0)
    ec5_ext_2 = ec_model.get_conductivity(T=temperature_range_extended,P=5,Cw=0.62,co2=0)
    dai_within  = pd.concat([pd.Series(ec5,index = temperature_range_3,name='Dai dry0.5'),
                             pd.Series(ec5_2,index = temperature_range_3,name='Dai dry_2gpa')],axis=1)
    dai_outside = pd.concat([pd.Series(ec5_ext_1,index = temperature_range_extended,name='Dai dry extr0.5'),
                             pd.Series(ec5_ext_2,index = temperature_range_extended,name='Dai dry xtr2gpa')],axis=1)
```

```python
within_experiments = np.log10(pd.concat([pom_df_within,ni_within,sif_within,dai_within],axis=1).sort_index())
extrapolated = np.log10(pd.concat([pom_df_outside,ni_outside,sif_outside,dai_outside],axis=1).sort_index())
```

```python
within_experiments.sort_index(inplace=True)
within_experiments
```

## Comparisons

```python
fig, ax = plt.subplots(figsize=(5,5),dpi=200)
P=2
for model in ['nih_11_basalt_hydrous', 'sifre_14_carb_h_basalt','dai_15_basalt']:
    if 'ni' in model:
        linestyle= '--'
    elif 'si' in model:
        linestyle='-'
    else:
        linestyle=':'
    ec_model_row = ecdatabase.get_model_properties(model)
    ec_model = ecdatabase.get_model(model)
    temperature_range = np.linspace(ec_model_row['temp_min'],
                                    ec_model_row['temp_max'],num=100,dtype=float).ravel()
    logfo2 = pyrutils.calc_QFM(temperature_range,P)
    for i, ppm_h2o in enumerate([0.5,1,3,5,7,10]):
        c = ec_model.get_conductivity(T=temperature_range,P=P,
                                               logfo2=logfo2,Cw=ppm_h2o*1e4,co2=0)
        ax.plot(1e4/temperature_range,
            np.log10(c),
            linewidth=2,linestyle=linestyle,color=cm.get_cmap('cool_r')(i/5))
        
for i, ppm_h2o in enumerate([0.5,1,3,5,7,10]):
    ax.plot([],[],linewidth=2,label=f' {ppm_h2o} Wt % Water',color=cm.get_cmap('cool_r')(i/5))
for model in ['nih_11_basalt_hydrous', 'sifre_14_carb_h_basalt','dai_15_basalt']:
    if 'ni' in model:
        linestyle= '--'
    elif 'si' in model:
        linestyle='-'
    else:
        linestyle=':'
    ax.plot([],[],linestyle=linestyle,color='black',label=model)
        
ax.fill_between(1e4/extrapolated.index, extrapolated.min(axis=1)-extrapolated.min(axis=1)*1e-3,extrapolated.max(axis=1),label='Extrapolated Dry Basalt',color='pink',alpha=0.5) 
ax.fill_between(1e4/within_experiments.index, within_experiments.min(axis=1)-within_experiments.min(axis=1)*1e-3,within_experiments.max(axis=1),label='In Bounds Dry Basalt',color='red',alpha=0.5)

ax2 = ax.twiny()
ticks = np.asarray([700,900,1200,1400,1600])
ax2.set_xticks(1e4/ticks,labels=ticks)
ax2.set_xlabel('T (K)',loc='left')
plt.legend(ncols=3)
ax.set_xlim([5,10])
ax.set_ylim([-5,0])

ax2.set_xlim([5,10])
ax.set_ylim([-4,2.5])
ax.legend(ncol=2,fontsize=7)
ax.set_title('Wet Basalt Conductivity Models')
ax.set_xlabel('1e4/T (K)',loc='left')
ax.set_ylabel('log10(sigma)')
fig.savefig('Wet Basalt Comparisons.png')
```
