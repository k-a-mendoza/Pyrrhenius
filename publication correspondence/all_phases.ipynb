```python
samples = 40
dfs = []

import numpy as np
import itertools

def generate_endpoint_combinations(conditions, N=10):
    sampled_conditions = {}
    endpoints =2**len(conditions.keys())
    if 1 > N/endpoints:
        N=3
    else:
        N = np.round(N/endpoints).astype(int)+1
    for key, values in conditions.items():
        # Convert values to numpy array for handling
        values = np.array(values)
        min_val = np.min(values)
        max_val = np.max(values)

        # Generate N subdivisions including the endpoints
        subdivisions = np.linspace(min_val, max_val, N)
        sampled_conditions[key] = subdivisions

    # Generate all combinations of the endpoints and subdivisions
    keys = list(sampled_conditions.keys())
    all_combinations = list(itertools.product(*(sampled_conditions[key] for key in keys)))

    # Create a new dictionary that maps each combination tuple to a list of arrays
    result_dict = {}
    for i, combination in enumerate(all_combinations):
        for j, key in enumerate(keys):
            if key not in result_dict:
                result_dict[key] = []
            result_dict[key].append(combination[j])

    # Convert lists to numpy arrays
    for key in result_dict:
        result_dict[key] = np.array(result_dict[key])

    return result_dict

for category, phases in phase_categories.items():
    
    for phase in phases:
        model = ecdatabase.get_model(phase)
        print(phase)
        conditions = model.generate_representative_conditions()
        new_conditions= generate_endpoint_combinations(conditions,N=samples)
        new_conditions['conductivity']=model.get_conductivity(**new_conditions)
        df = pd.DataFrame(new_conditions)
        if 'Cw' not in conditions.keys() and model.uses_water:
            df['Cw']=model.metadata.convert2ppm(model.metadata.water_average)
        if 'X_fe' not in conditions.keys() and model.uses_iron:
            df['X_fe']=model.metadata.iron_average

        if 'P' not in conditions.keys() and model.uses_pressure:
            df['X_fe']=model.metadata.pressure_average
        
        df['category']=category.replace('_',' ').title()
        df['model']=model
        df['mineral']=model.metadata.phase_type 
        dfs.append(df)

df = pd.concat(dfs,ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 4), dpi=200)
ax.set_xscale('log')
df['category'] = df['category'].replace({
    'Mtz Phases': 'MTZ Minerals',
    'Crustal Rocks Minerals Igneous Metamorphic': 'Crustal Rocks',
    'Upper Mantle Rocks Minerals': 'Upper Mantle Minerals'
})
# Calculate min and max for each category and plot as horizontal bars
categories = ['Crustal Rocks', 'Upper Mantle Minerals','MTZ Minerals', 
       'Aqueous Fluids', 'Magmas']
color_map = {
    'MTZ Minerals': 'goldenrod',
    'Crustal Rocks': 'slategray',
    'Upper Mantle Minerals': 'sienna',
    'Aqueous Fluids': 'steelblue',
    'Magmas': 'firebrick'
}
for category in categories:
    subset = df[df['category'] == category]['conductivity']
    min_val = subset.min()
    max_val = subset.max()
    ax.hlines(y=category, xmin=min_val, xmax=max_val, color='black', linewidth=22)
    ax.hlines(y=category, xmin=min_val*1.05, xmax=max_val*0.95, color=color_map[category], linewidth=20)

# Set axis limits and labels
ax.set_xlim([1e-5, 1e3])
ax.set_xlabel('Conductivity (S/m)')

# Add secondary axis for log resistivity
def forward(x):
    return 1 / x

def inverse(x):
    return 1 / x

secax = ax.twiny()
secax.set_xlabel('Resistivity (m)')
secax.set_xlim(1e5, 1e-3)  # Set the resistivity limits directly
secax.set_xscale('log')
# Custom ticks based on expected resistivity values
resistivity_ticks = [1e5, 1e4, 1e3, 1e2, 1e1, 1, 1e-1, 1e-2, 1e-3]
secax.set_xticks(resistivity_ticks)

ax.grid(axis='x')
# Remove any existing legends (not needed here)

# Improve the layout and aesthetics
plt.tight_layout()
plt.savefig('Rock_Conditions.pdf')
plt.show()
```
